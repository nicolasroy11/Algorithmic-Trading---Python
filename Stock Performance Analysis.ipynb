```python
# normalizing to compare performances

# this line takes the closing prices and divides all of them by their very first value (.iloc[0] says "row 0 
# and whatever column", so naturally, the first value is always 1, beacuse it is
# the first value divided by itself), then it multiplies all the values by 100
# if I wanted to only do this vectorized operation to, say, ETH-USD, I would write:
# closing_prices['ETH-USD'].div(closing_prices.iloc[0, 0]).mul(100). This would cause every value in the ETH-USD
# column to get divided by the very first value of the column; using .iloc[0, :] is just a generalization
# of the vectorized operation to the whole table. This also gives the different assets the same starting point
normalized_closing_prices = closing_prices.div(closing_prices.iloc[0]).mul(100)
normalized_closing_prices.head()
```

```python
normalized_closing_prices.plot(figsize=(15, 8), fontsize = 12)
plt.show()
```

```python
# now lets make a new df with percent movements for each period. Instead of an absolute price, we will have percent growth
# since the last price

# first, we make the two values we will compare
DOGE = closing_prices['DOGE-USD'].copy().to_frame()
DOGE['lagged_by_one'] = DOGE.shift(periods = 1)
DOGE.head()
```

```python
# then we carry out the vectorized division
# we subtract 1 (.sub(1)) because the value is a only quotient. We need to see a difference from one period to the next
DOGE['percent_growth'] = DOGE['DOGE-USD'].div(DOGE['lagged_by_one']).sub(1).mul(100)
DOGE[0:10]
```

```python
# but here is the smart way to do it:
DOGE['percent_growth'] = DOGE['DOGE-USD'].pct_change(periods = 2).mul(100)
DOGE.head(10)
```

```python
# here we will resample the df. We will create a df that only has the last value of each day in the original dataframe
# and get the percent change from that to get an idea of how much change the asset underwent each day.
# we drop N/A entries
resampled_DOGE = DOGE['DOGE-USD'].resample('D').last().pct_change(periods = 1 ).mul(100).dropna()
resampled_DOGE.head(10)
```

# Analyzing performance


```python
import numpy as np
```

```python
# starting over with the Doge closing prices; calculating returns (% percent change)
doge = closing_prices['DOGE-USD'].copy().to_frame()
doge["returns"] = closing_prices['DOGE-USD'].pct_change(periods = 1).mul(100).dropna()
doge["returns"].plot(kind = "hist", figsize = (12, 8), bins = 100)
```

```python
# not surprisingly, the majority of returns center around 0
# There are 2 metrics to use in summarizing these distributions
# and trying to assess whether an asset is worth plugging at: value and volatility
# - value: the mean return (the higher the mean return, the more profitable)
# - volatility: the std_dev of the returns about the mean (too much variance means too much volatility, which implies a rather large risk)

mean_return = doge["returns"].mean()
mean_return * 100 # the mean percent return, negative in this case, indicating poor returns during this period
```

```python
# now we calculate the std_dev to see how volatile the asset is
dev_return = doge["returns"].std()
dev_return * 100 # a high deviation, likely an artifact of the sudden drop 9-8-2021
```

Now let's do a comparative analysis of the 3 coins

```python
# let's get ~2 months' worth of price data
tickers = ['ETH-USD', 'LTC-USD', 'DOGE-USD']
coins = yf.download(tickers, start = "2021-08-16", end = "2021-09-13", interval='5m')
```

```python
# grab the closing prices column
close = coins['Close']

# calculate the % change by 1 shift
returns = close.pct_change(periods = 1)

# take a look at the stats summary for these series
returns.describe()
```

```python
returns.describe().T # we can transpose the above matrix to invert the rows/columns
```

```python
returns.describe().T.loc[:, ["mean", "std"]] # we can also reduce to the only data we need
```

```python
import matplotlib.pyplot as plt

# we'll save it as its own dataframe
summary = returns.describe().T.loc[:, ["mean", "std"]]

# we can now throw up a scatter plot to visualize the risk/return (those in the leftmost and topmost positions win)
summary.plot.scatter(x = "std", y = "mean")
plt.xlabel("Risk (std)")
plt.ylabel("Return (mean)")
plt.title("Risk/Return", fontsize=15)

# put the names of the corresponding asset next to the points
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"] + 0.0000002, summary.loc[i, "mean"] + 0.0000002))

plt.show()
```

# Simple Moving Average with rolling()

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# we get the closing prices of the returns
DOGE = closing_prices['DOGE-USD'].copy().to_frame()
DOGE.head()
```

```python
DOGE.plot()
plt.legend(loc = "upper left", fontsize = 15)
plt.show()
```

```python
# generate a moving window for this series for 10 candles
# note that the window's purpose is not yet defined
# what do we do with these groupings of 10? How are these windowed values going to be aggregated?
DOGE.rolling(window = 10)
```

```python
# this returns a Pandas object called a Rolling
type(DOGE.rolling(window = 10))
```

```python
# well, we could get the mean every 10 candles by chaining mean() to the Rolling object
# naturally we will expect the first 9 values to be undefined.
# here is your basic SMA:
DOGE.rolling(window = 10).mean()
```

```python
# or the median
DOGE.rolling(window = 10).median()
```

```python
# you could also do the min value
DOGE.rolling(window = 10).min()
```

```python
# you could also specify min_periods, which will start giving you values after 5 non-NA values
# despite the windowing being of 10 periods. Then you'll start getting values at 5 periods (candles),
# will not be averaging more than 10 candles per window.
DOGE.rolling(window = 10, min_periods = 5).mean()
```
